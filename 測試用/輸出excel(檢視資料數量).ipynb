```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

資料來源：https://huggingface.co/datasets/common_voice/blob/main/dataset_infos.json

```python
with open("dataset_infos.json", "r", encoding="utf-8") as file:
    data = json.load(file)
```

```python
# 獲取詳細數量
result =  {lang : 
           {split : data[lang]["splits"][split]["num_examples"] for split in data[lang]["splits"].keys()} 
           for lang in data.keys()}
```

```python
result["ab"]
```

```python
df = pd.DataFrame(result)

# 輸出成 excel
df.T.to_excel("各語言資料詳細數量.xlsx")
```
